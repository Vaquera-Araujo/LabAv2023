# Canonical Equations and Poisson Brackets

## Preamble

```python
from sympy import symbols, functions, Function
```

```python
q,p = symbols('q,p')
f = Function('f')(q,p)
g = Function('g')(q,p)
h = Function('h')(q,p)
```

## Poisson bracket for 2D phase space

```python
def Pb(f,g):

    return f.diff(q)*g.diff(p) - f.diff(p)*g.diff(q)
```

```python
Pb(f,g)
```

##  Jacobi Identity

```python
(Pb(f,Pb(g,h))+Pb(g,Pb(h,f))+Pb(h,Pb(f,g))).expand()
```
